## What is R Squared ($R^2$)?

- **R Squared** is a statistical measure that shows how well the regression predictions approximate the real data points.
- It represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).
- **Range:** $0 \leq R^2 \leq 1$ (or 0% to 100%)
    - $R^2 = 1$: Perfect fit (model explains all variance)
    - $R^2 = 0$: Model explains none of the variance

### Formula

$$
R^2 = 1 - \frac{SS_{res}}{SS_{total}}
$$

Where:
- $SS_{res} = \sum (y_i - \hat{y}_i)^2$ (Sum of squared residuals/errors)
- $SS_{total} = \sum (y_i - \bar{y})^2$ (Total sum of squares)
- $y_i$: Actual value
- $\hat{y}_i$: Predicted value
- $\bar{y}$: Mean of actual values

### How to Interpret

- $R^2 = 0.70$ means the model explains 70% of the variance in the data.
- Higher $R^2$ means a better fit, but very high values can indicate overfitting.

## What is Adjusted R Squared ($R^2_{adj}$)?

- **Adjusted R Squared** adjusts the $R^2$ value for the number of predictors in the model.
- It penalizes the addition of unnecessary independent variables.
- Useful when comparing models with different numbers of predictors.

### Formula

$$
R^2_{adj} = 1 - \left(1 - R^2\right) \frac{N-1}{N-p-1}
$$

Where:
- $N$: Number of data points
- $p$: Number of independent features

### Why Use Adjusted R Squared?

- Adding more variables to a model can artificially increase $R^2$, even if those variables are not meaningful.
- Adjusted $R^2$ only increases if the new variable improves the model more than would be expected by chance.

---

## Key Points

- **$R^2$** tells you how well your model fits the data.
- **Adjusted $R^2$** tells you how well your model fits, accounting for the number of predictors.
- Use **Adjusted $R^2$** when comparing models with different numbers of features.
