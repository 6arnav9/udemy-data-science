## 1Ô∏è‚É£ Equation of a Line (2D)

The most basic form:
\[
y = mx + c
\]

- **m:** Slope (steepness of the line)
- **c:** Y-intercept (where the line crosses the y-axis)

### Vector Form:
\[
y = \beta_0 + \beta_1 x
\]

Or rearranged:
\[
ax + by + c = 0
\]

---

## 2Ô∏è‚É£ Equation of a Plane (3D)

In 3D, the general equation of a plane is:
\[
w_1 x_1 + w_2 x_2 + w_3 x_3 + b = 0
\]

### Compact Vector Form:
\[
w^T x + b = 0
\]

Where:
- \( w = \begin{bmatrix} w_1 \\ w_2 \\ w_3 \end{bmatrix} \) ‚Üí Weight (normal) vector
- \( x = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix} \) ‚Üí Position vector

---

## 3Ô∏è‚É£ Hyperplane in n-Dimensions

The equation extends to n dimensions:
\[
w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b = 0
\]

### Compact Vector Form:
\[
w^T x + b = 0
\]

---

## 4Ô∏è‚É£ Special Case: Hyperplane Passing Through the Origin

When the hyperplane passes through the origin:
\[
w^T x = 0
\]

### Key Points:
- **w:** Always perpendicular (normal) to the hyperplane.
- **x:** Any point lying on the hyperplane.
- The dot product \( w^T x = |w||x|\cos \theta \)
- For the equation to hold: \( \cos \theta = 0 \) ‚Üí \( \theta = 90^\circ \)
- This means **w and x are perpendicular only for points on the hyperplane when passing through the origin.**

---

## üîç Important Concept: Perpendicularity

- **w is always perpendicular to the hyperplane.**
- **w is NOT always perpendicular to all x vectors.**
- **w is perpendicular only to those x vectors that satisfy the hyperplane equation.**
- For the origin case (b = 0), the condition is \( w^T x = 0 \) ‚Üí perpendicularity.

## ‚úÖ Key Takeaways

- The **hyperplane** divides the space into two parts.
- The **w vector is always perpendicular to the hyperplane.**
- The **x vectors** are position vectors. They may or may not lie on the hyperplane.
- The **dot product tells us** if a point lies on the hyperplane.
- When the hyperplane passes through the origin ‚Üí **the dot product is zero** ‚Üí w is perpendicular to x.
