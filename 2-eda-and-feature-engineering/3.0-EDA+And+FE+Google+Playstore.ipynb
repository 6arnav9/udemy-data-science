 ## ðŸ“Š Common EDA Graphs for Categorical & Numerical Features

### 1. Categorical Features

| Plot Type      | Use Case / What It Shows                                 |
|----------------|---------------------------------------------------------|
| **Bar Plot**   | Frequency/count of each category                        |
| **Count Plot** | Like bar plot, but directly from categorical data       |
| **Pie Chart**  | Proportion of each category (good for few categories)   |
| **Box Plot**   | Distribution of a numerical variable across categories  |
| **Stacked Bar**| Compare two categorical variables                       |



---

### 2. Numerical Features

| Plot Type        | Use Case / What It Shows                                  |
|------------------|----------------------------------------------------------|
| **Histogram**    | Distribution (shape, skewness, outliers)                 |
| **KDE Plot**     | Smoothed distribution (density)                          |
| **Box Plot**     | Quartiles, median, outliers                              |
| **Violin Plot**  | Distribution + density + outliers                        |
| **Scatter Plot** | Relationship between two numerical features              |
| **Pair Plot**    | Relationships between all pairs of numerical features    |




---

### 3. Mixed (Categorical vs Numerical)

| Plot Type        | Use Case / What It Shows                                  |
|------------------|----------------------------------------------------------|
| **Box Plot**     | Distribution of numerical variable by category            |
| **Violin Plot**  | Distribution + density by category                        |
| **Swarm Plot**   | All data points by category (good for small datasets)     |
| **Bar Plot**     | Aggregate (mean/sum) of numerical by category             |




---

### 4. Correlation & Relationships

| Plot Type        | Use Case / What It Shows                                  |
|------------------|----------------------------------------------------------|
| **Heatmap**      | Correlation matrix between numerical features             |
| **Pair Plot**    | Scatterplots for all pairs of numerical features          |






---

> **Tip:**  
> - Use **bar/count/pie** for categorical features.  
> - Use **histogram/KDE/box/violin** for numerical features.  
> - Use **box/violin/bar** for categorical vs numerical.  
> - Use **scatter/pair/heatmap** for relationships and correlations.

```python
# Some more data cleaning, not in the notebook originally provided

# Check the number of missing values in each column
df_copy.isnull().sum()

# Display rows where 'Type' is null
df_copy[df_copy['Type'].isnull()]

# Drop rows where 'Type' is null
df_copy.drop(df_copy[df_copy['Type'].isnull()].index, axis=0, inplace=True)

# Another way to drop rows with null 'Type' (commented out)
# df_copy = df_copy[df_copy['Type'].notnull()]

# Confirm that no rows with null 'Type' remain
df_copy[df_copy['Type'].isnull()]

# Calculate the median for 'Rating' and 'Size' columns
rating_median, size_median = df_copy['Rating'].median(), df_copy['Size'].median()

# Fill missing values in 'Rating' and 'Size' with their respective medians
df_copy['Rating'].fillna(rating_median, inplace=True)
df_copy['Size'].fillna(size_median, inplace=True)

# Check remaining missing values after imputation
df_copy.isnull().sum()

# Get the most frequent value (mode) for 'Current Ver' and 'Android Ver'
currentVer_mode = df_copy['Current Ver'].mode()[0]
androidVer_mode = df_copy['Android Ver'].mode()[0]
print(currentVer_mode, androidVer_mode)

# Fill missing values in 'Current Ver' and 'Android Ver' with their modes
df_copy['Current Ver'].fillna(currentVer_mode, inplace=True)
df_copy['Android Ver'].fillna(androidVer_mode, inplace=True) 

# Final check for any remaining missing values
print(df_copy.isnull().sum())
```

## Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2, palette='hls')
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store

## What are the Top 5 most installed Apps in Each popular Categories ??

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
print(dfa)
apps = ['GAME', 'COMMUNICATION', 'TOOLS', 'PRODUCTIVITY']
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App', palette='hls')
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.

## How many apps are there on Google Play Store which get 5 ratings??

```python
rating = df_copy.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()
# rating = df_copy.groupby(['App'])['Rating'].sum().sort_values(ascending = False).reset_index()
print(rating)

toprating_apps = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```
